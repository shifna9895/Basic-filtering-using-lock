```python
import pandas as pd
import numpy as np
path="/content/weather_data.csv"
df=pd.read_csv(path)
df
```

```python
# 1) filter the dataframe where temerature is>30
df[df['temperature']>30]

# to print the dataframe if the condition is true  df[df['specify item']condition]
```

```python
# 2) find temperature is >30
df['temperature']>30
```

```python
# 3) find dataframe where windspeed<=2
df[df['windspeed']<=2]
```

```python
# 4) a) find the day where wind speed is<=5
df[df['windspeed']<=5].day
```

```python
# 4) b)
df['day'][df['windspeed']<=5]
# to retrieve specified items in a data columns  df[item]df[column name]condition]
```

```python
# 5) find the day and temperature where wid speed<=7
df[['day','temperature']][df['windspeed']<=7]
```

```python
# 6) find the temperature and windspeed where the temperature<=30
df[['temperature','windspeed']][df['temperature']<=30]
```

```python
# 7) find the day in which maximum temperatue recorded
max=df['temperature'].max()
df['day'][df['temperature']==max]
```

```python
# 7) b)
df['day'][df['temperature']==df['temperature'].max()]
```

```python
# 8) find the day and temperature in which the minimum temperature recorded
df[['day','temperature']][df['temperature']==df['temperature'].min()]
```

```python
# 9) find the days in which the temperature went above average
df['day'][df['temperature']>=df['temperature'].mean()]
```

```python
# 10) find either temperature less than 20 or windspeed greater than 5
df[(df['temperature']<20) | (df['windspeed']>5)]
```

```python
# 11) find event snow & temperature less than 30
df[(df['event']=='Snow') & (df['temperature']<30)]
```

```python
# 12) temperature greater than 25  and windspeed less than 5 or temperature less than 25 and wind speed greater than 5
df[((df['temperature']>25)&(df['windspeed']<5))| ((df['temperature']<30)&(df['windspeed']>5))]
```

```python
# 13) find dataframe where wind speed>1 and event=snow
df[(df['windspeed']>1)&(df['event']=='Snow')]
```

```python
# 14) find dataframe where wind speed<5 and event is not equal to rain
df[(df['windspeed']<5)&(df['event']!='Rain')]
```

```python
# 15) find dataframe if windspeed>3 and event=snow or windspeed<3 and event!=.rain
df[((df['windspeed']>3)&(df['event']=='Snow')|(df['windspeed']<3)&(df['event']!='Rain'))]
```

```python
# 16) find day in which temperature>20 or windspeed<5
df['day'][(df['temperature']>20)|(df['windspeed']<5)]
```

```python
# 17) find event, day where temperature<20 or windspeed>6
df[['event','day']][(df['temperature']<20)|(df['windspeed']>6)]
```

```python
# 18) find day,event in which temperature>30 or windspeed<4 and temperature<20 or windspeed>5
df[['day','event']][((df['temperature']>30)|(df['windspeed']<4))&((df['temperature']<20)|(df['windspeed']>5))]
```

```python
# 19) find temperature,event where temperature>=30 and windspeed>=4
df[['temperature','event']][(df['temperature']>=30)&(df['windspeed']>=4)]
```

```python
# 20) find event where temperature <=28 and windspeed<=4
df['event'][(df['temperature']<=28)&(df['windspeed']<=4)]
```

```python
# 21) find temperature,event where temperature>=35 and windspeed>=5 or temperature<=28and windspeed<=5
df[['temperature','event']][((df['temperature']>=35)&(df['windspeed']>=5))|((df['temperature']<=28)&(df['windspeed']<=5))]
```

```python
# 22) find dataframe in which event=Rain
df[(df['event']=='Rain')]
```
