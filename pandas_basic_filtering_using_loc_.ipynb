```python
import pandas as pd
import numpy as np
```

```python
path="/content/Titanic-Dataset.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
# 1) Get the details of passengers who survived and were younger than 15 years old.
df.loc[(df['Survived']==1)&(df['Age']<15)]
```

```python

# 2) Get the details of male passengers who were older than 35 years old.
df.loc[(df['Sex']=='male')&(df['Age']>35)]
```

```python
# 3) Get the details of passengers who paid a fare greater than $40 and survived.
df.loc[(df['Fare']>40)&(df['Survived']==1)]
```

```python
# 4) Get the details of male passengers who did not survive.
df.loc[(df['Sex']=='male')&(df['Survived']==1)]
```

```python
# 5) Get the details of passengers who were in either the first or second class.
df.loc[(df['Pclass']==1)|(df['Pclass']==2)]
```

```python
# 6) Get the details of passengers who had more than two neighbors/spouses aboard the Titanic and did not survive.
df.loc[(df['SibSp']>2)&(df['Survived']==0)]
```

```python
# 7) Get the details of passengers who were in the third class and were older than 25 years old.
df.loc[(df['Pclass']==3)&(df['Age']>25)]
```

```python
# 8) Get the details of passengers who were younger than 18 years old and had parents/children aboard.
df.loc[(df['Age']<18)&(df['Parch']>0)]
```

```python
# 9) Get the details of male passengers who paid less than $15 for their fare.
df.loc[(df['Sex']=='male')&(df['Fare']<15)]
```

```python
# 10) Get the details of passengers who were in the third class, younger than 18, and had siblings/spouses aboard.
df.loc[(df['Pclass']==3)&(df['Age']<18)&(df['SibSp']>0)]
```

```python
# 11) Get the details of passengers who were either younger than 5 years or older than 55 years, and survived.
df[((df['Age']<5)|(df['Age']>55))&(df['Survived']==1)]
```

```python
# 12) get the name, passenger id and ticket no. who were female and embarked in southampton
female_passengers_southampton=df.loc[(df['Sex']=='female')&(df['Embarked']=='S'),['Name','PassengerId','Ticket']]
female_passengers_southampton
```

```python
# 13) get the names of female pssengers who were survived.
survived_female_names=df.loc[(df['Sex']=='female')&(df['Survived']==1),'Name']
survived_female_names
```

```python
# 14) get the name, passenger id and cabin no. who were in the second class and the ticket fare greater than 60
details_secondcls_faregreater100=df.loc[(df['Pclass']==2)&(df['Fare']>60),['Name','PassengerId','Cabin']]
details_secondcls_faregreater100
```

```python
# 15) get the details of passengers who were embarked in cherbourg or older than 40 years
dp_embarked_c_age_greater40=df.loc[(df['Embarked']=='C')|(df['Age']>40)]
dp_embarked_c_age_greater40
```

```python
# 16) get the details of passengers in between 1 to 10 years old
child=df.loc[(df['Age']>1)&(df['Age']<10)]
child
```

```python
# 17) get the details of passengers index  500 to 650
df.loc[500:650]
```

```python
# 18) get the details of passengers who have siblings and spouses abroad is more than 2 or parents and children abroad is less than 2.
sibsp_parch=df.loc[(df['SibSp']>2)&(df['Parch']<2)]
sibsp_parch
```

```python
# 19) get the details of female passengers who were embarked in southampton.
female_s=df.loc[(df['Sex']=='female')&(df['Embarked']=='S')]
female_s
```

```python
# 20) get the name, age, ticket, cabin no and fare of male passengers who were embarked in queenstown.
male_q=df.loc[(df['Sex']=='male')&(df['Embarked']=='Q'),['Name','Age','Ticket','Cabin','Fare']]
male_q
```

```python
# 21) get the details of female passengers who have more than 3 siblings and spouses abroad.
female_sibsp=df.loc[(df['Sex']=='female')&(df['SibSp']>3)]
female_sibsp
```

```python
# 22) get the male passenger details who either younger than 18 and older than 70.
male_age_bw_18_70=df.loc[(df['Sex']=='male')&((df['Age']<18)|(df['Age']>70))]
male_age_bw_18_70
```

```python
# 23) get the details of passengers who were in second class, older than 50 and parents/children abroad.
secondcls_age50_parch=df.loc[(df['Pclass']==2)&(df['Age']>50)&(df['Parch']>0)]
secondcls_age50_parch
```

```python
# 24) get the details of passengers in first class index 2:45.
first_cls_passenger=df.loc[(df.index>=2)&(df.index<=45)&(df['Pclass']==1)]
first_cls_passenger
```

```python
# 25) get the passenger id,name and class who where embarked in queenstown and survived index 100 to 700.
q_survivals=df.loc[(df.index>=100)&(df.index<=700)&(df['Embarked']=='Q')&(df['Survived']==1),['Name','PassengerId','Pclass']]
q_survivals
```
